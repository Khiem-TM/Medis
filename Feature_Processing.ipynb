```python
import sqlite3
import pandas as pd

path = "/content/drive/MyDrive/DDIMDL/event.db"
connect = sqlite3.connect(path)
query_tables = "SELECT name FROM sqlite_master WHERE type='table';"
tables = pd.read_sql_query(query_tables, connect)
print("Các bảng hiện có trong database:")
print(tables)
```

```python
df_event_number = pd.read_sql("SELECT * FROM event_number", connect)
df_event = pd.read_sql("SELECT * FROM event", connect)
df_drug = pd.read_sql("SELECT * FROM drug", connect)
df_extraction = pd.read_sql("SELECT * FROM extraction", connect)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
df_plot = df_event_number.sort_values(by='number', ascending=False)

# ==========================================
# BIỂU ĐỒ 1: TOP 15 TƯƠNG TÁC PHỔ BIẾN NHẤT
# ==========================================
plt.figure(figsize=(12, 8))
sns.barplot(x='number', y='event', data=df_plot.head(15), palette='Reds_r')

plt.title('Top 15 loại tương tác thuốc xuất hiện nhiều nhất', fontsize=16, fontweight='bold')
plt.xlabel('Số lượng cặp thuốc (Cỡ mẫu)', fontsize=12)
plt.ylabel('Loại tương tác (Event)', fontsize=12)
plt.tight_layout()
plt.show()

# ==========================================
# BIỂU ĐỒ 2: ĐỘ DỐC CỦA TOÀN BỘ 65 CLASS
# ==========================================
plt.figure(figsize=(14, 5))
plt.plot(range(len(df_plot)), df_plot['number'], marker='o', markersize=4, color='b', linestyle='-')
plt.fill_between(range(len(df_plot)), df_plot['number'], color='blue', alpha=0.1)

plt.title('Phân phối tổng quan 65 Class: Hiện tượng "Cái đuôi dài" (Long-tail)', fontsize=16, fontweight='bold')
plt.xlabel('Thứ hạng Class (Từ phổ biến nhất -> Hiếm nhất)', fontsize=12)
plt.ylabel('Số lượng mẫu', fontsize=12)
plt.axhline(y=100, color='r', linestyle='--', label='Ngưỡng 100 mẫu')
plt.legend()
plt.tight_layout()
plt.show()
```
